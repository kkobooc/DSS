```python
from matplotlib.image import imread
image = imread('./data/ladybug.png')
image.shape
```

```python
plt.imshow(image);
```

```python
from sklearn.cluster import KMeans

X = image.reshape(-1, 3)
kmeans = KMeans(n_clusters=8, random_state=13).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img);
```

```python
segmented_imgs = []
n_colors = (10, 8, 6, 4, 2)
for n_clusters in n_colors:
    kmeans = KMeans(n_clusters=n_clusters, random_state=13).fit(X)
    segmented_img = kmeans.cluster_centers_[kmeans.labels_]
    segmented_imgs.append(segmented_img.reshape(image.shape))
```

```python
plt.figure(figsize=(10,5))
plt.subplots_adjust(wspace=0.05, hspace=0.1)

plt.subplot(231)
plt.imshow(image)
plt.title("Original image")
plt.axis('off')

for idx, n_clusters in enumerate(n_colors):
    plt.subplot(232 + idx)
    plt.imshow(segmented_imgs[idx])
    plt.title('{} colors'.format(n_clusters))
    plt.axis('off')

plt.show()
```

### MNIST로 실습

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

X_digits, y_digits = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits, random_state=13)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(multi_class="ovr", solver="lbfgs", max_iter=5000, random_state=13)
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
# Pipeline 통과
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('kmeans', KMeans(n_clusters=50, random_state=13)),
    ('log_reg', LogisticRegression(multi_class='ovr',
                                  solver='lbfgs', max_iter=5000, random_state=13))
])
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

```python
# GridSearch
from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=range(2, 100))
gird_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)
gird_clf.fit(X_train, y_train)
```

```python
gird_clf.best_params_
```

```python
gird_clf.score(X_test, y_test)
```
