### 회귀(Regession)

```python
import pandas as pd

data = {'x' : [1., 2., 3., 4., 5.], 'y':[1., 3., 4., 6., 5.]}
df = pd.DataFrame(data)
df
```

```python
import statsmodels.formula.api as smf
# statsmodel
# 다양한 통계 모델을 평가하고, 통계 테스트를 수행하고, 통계 데이터를 탐색하는데 필요한 모듈

# ols -> 최소제곱법을 이용한 회귀 모형 작성하는 함수
# formula = '종속변수 ~ 독립변수'
lm_model = smf.ols(formula='y ~ x', data=df).fit()
```

```python
# 결과 확인
lm_model.params
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lmplot(x='x', y='y', data=df);
# 진한 영역은 각 값의 신뢰구간, 0~100의 값을 ci= 에서 정할 수 있다
```

```python
sns.regplot(x='x', y='y', data=df);
```

```python
# 잔차 확인
resid = lm_model.resid
resid
```

```python
# numpy로 직접 결정계수 계싼
import numpy as np

mu = np.mean(df.y)
y = df.y
yhat = lm_model.predict()
np.sum((yhat-mu)**2 / np.sum((y-mu)**2))
```

```python
# statsmodel의 r2 함수로 쉽게 구학
lm_model.rsquared
```

```python
sns.distplot(resid, color='black');
```
