### 실험용 데이터 읽어오기

```python
import numpy as np

raw_data = np.genfromtxt('./data/x09.txt', skip_header=36)
raw_data
```

### 데이터 시각화

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

xs = np.array(raw_data[:,2], dtype=np.float32)
ys = np.array(raw_data[:,3], dtype=np.float32)
zs = np.array(raw_data[:,4], dtype=np.float32)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```

```python
# 학습 데이터 추리기
x_data = np.array(raw_data[:, 2:4], dtype=np.float32)
y_data = np.array(raw_data[:, 4], dtype=np.float32)

y_data = y_data.reshape((25, 1))
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2, )),
])

model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.summary()
```

```python
hist = model.fit(x_data, y_data, epochs=5000)
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
# loss가 잘 떨어진다
```

```python
# 가중치와 bias 확인
W_, b_ = model.get_weights()
print('Weigh is : ', W_)
print('bias is : ', b_)
```

```python
x = np.linspace(20, 100, 50).reshape(50, 1)
y = np.linspace(10, 70, 50).reshape(50, 1)

X = np.concatenate((x, y), axis=1)
Z = np.matmul(X, W_) + b_
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.scatter(x, y, Z)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```

### XOR 문제

```python
# 실습용 데이터 준비
X = np.array([[0, 0],
              [1, 0],
              [0, 1],
              [1, 1] ])
y = np.array([[0], [1], [1], [0]])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, activation='sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# 옵티마이저를 SGD로, 학습률을 0.1 로 지정
# loss 함수는 mse로
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1), loss='mse')
```

```python
model.summary()
```

```python
# 학습
hist = model.fit(X, y, epochs=5000, batch_size=1)
```

```python
model.predict(X)
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```

```python
# 학습에서 찾은 가중치
for w in model.weights:
    print('---------------------------------')
    print(w)
```

### 아이리스 데이터로 딥러닝 분류

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target
```

```python
y
```

```python
# 타겟 데이터 원핫인코딩
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
enc.fit(y.reshape(len(y), 1))
```

```python
enc.categories_
```

```python
y_onehot = enc.transform(y.reshape(len(y), 1))
y_onehot
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot,
                                                   test_size=0.2,
                                                   random_state=13)
```

```python
# 모델 구성
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(4, input_shape=(4,), activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')    
])
```

```python
# optimizer = adam
# loss함수 = categorical_crossentropy
# metrics = accuracy

model.compile(optimizer='adam', loss='categorical_crossentropy',
             metrics=['accuracy'])
model.summary()
```

```python
hist = model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test, verbose=2)
```
