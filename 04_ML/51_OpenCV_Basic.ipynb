## openCV

```python
import matplotlib.pyplot as plt
import cv2
```

```python
imgBGR = cv2.imread('./data/sin.jpg')
plt.axis('off')
plt.imshow(imgBGR)
plt.show()
```

```python
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(imgRGB)
plt.show()
```

```python
imgGray = cv2.imread('./data/sin.jpg', cv2.IMREAD_GRAYSCALE)
plt.axis('off')
plt.imshow(imgGray, cmap='gray')
plt.show()
```

### 마스크 연산

```python
src = cv2.imread('./data/airplane.bmp', cv2.IMREAD_COLOR)
plt.axis('off')
plt.imshow(src);
```

```python
mask = cv2.imread('./data/mask_plane.bmp', cv2.IMREAD_GRAYSCALE)
plt.axis('off')
plt.imshow(mask, cmap='gray');
```

```python
dst = cv2.imread('./data/field.bmp', cv2.IMREAD_COLOR)
plt.axis('off')
plt.imshow(dst);
```

```python
cv2.copyTo(src, mask, dst=None)
```

```python
plt.axis('off')
plt.imshow(cv2.copyTo(src, mask, dst));
```
