```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color="0.5")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

```python
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"],
            y_vars=["petal_width", "petal_length"],
            hue="species",
            height=4)
plt.show()
```
